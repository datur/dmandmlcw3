```python
import tensorflow as tf
import pandas as pd
import keras

from keras import Sequential
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation

from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('fer2017-training.csv')
```

```python
train_data.head()
```

```python
train_labels = train_data['emotion'].copy()

train_data.drop('emotion',axis = 1, inplace=True)
```

```python
train_n_samples, n_features = train_data.shape
n_classes = len(set(train_labels))
```

```python
X_train = train_data.values
y_train = train_labels.values
```

```python
test_data = pd.read_csv('fer2017-testing.csv')
```

```python
test_data.head()
```

```python
test_labels = test_data['emotion'].copy()

test_data.drop('emotion',axis = 1, inplace=True)
```

```python
test_n_samples, n_features = test_data.shape
```

```python
X_test = test_data.values
y_test = test_labels.values
```

```python
#normalise data

X_test = X_test/255
X_train = X_train/255
```

```python
#onehotlabels

y_train_labels = keras.utils.to_categorical(y_train, num_classes=n_classes)
y_test_labels = keras.utils.to_categorical(y_test, num_classes=n_classes)
```

```python
model = Sequential()

#input
model.add(Dense(n_features, input_shape=X_train[0].shape, activation='relu'))

# hidden layers
model.add(Dense(64, activation='relu'))

model.add(Dense(64, activation='relu'))

model.add(Dense(64, activation='relu'))


#output
model.add(Dense(n_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train_labels, 
          epochs=20, 
          verbose=1, batch_size=16,
          validation_data=(X_test, y_test_labels))
```

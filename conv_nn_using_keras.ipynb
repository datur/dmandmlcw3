```python
import tensorflow as tf
import pandas as pd
import keras
from keras import Sequential
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D

from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('fer2018.csv')
```

```python
data.head()
```

```python
labels = data['class'].copy()

data.drop('class',axis = 1, inplace=True)
```

```python
#labels
```

```python
n_samples, n_features = data.shape
n_classes = len(set(labels))
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

```

```python
#for i in range(5):
#    plt.matshow(img_data[i])
#plt.show()
```

```python
X_train_vals = X_train.values
X_test_vals = X_test.values

y_test_vals = y_test.values
y_train_vals = y_train.values
```

```python
for i in X_train_vals:
    print(i)
```

```python
img_data = []
x_test_data = []
y_train_data = []
y_test_data = []
for i in range(len(X_train_vals)):
    img_data.append((X_train_vals[i]/255).reshape(48,48))
for i in range(len(X_test_vals)):
    x_test_data.append((X_test_vals[i]/255).reshape(48,48))
for i in range(len(y_train_vals)):
    y_train_data.append(y_train_vals[i])
for i in range(len(y_test_vals)):
    y_test_data.append(y_test_vals[i])
```

```python
X_train_vals = np.array(img_data)
X_test_vals = np.array(x_test_data)
y_train_vals = np.array(y_train_data)
y_test_vals = np.array(y_test_data)

```

```python
plt.imshow(X_train_vals[0])
```

```python
X_train_vals_final = X_train_vals.reshape(28709, 48, 48,1)
```

```python
X_test_vals.shape
```

```python
X_test_vals_final = X_test_vals.reshape(7178, 48, 48,1)
y_train_vals_final = keras.utils.np_utils.to_categorical(y_train_vals, n_classes)
y_test_vals_final = keras.utils.np_utils.to_categorical(y_test_vals, n_classes)
```

```python

model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5), strides=(1, 1),
                 activation='relu',
                 input_shape=(48, 48,1)))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dense(7, activation='softmax'))


model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
batch_size = 128
epochs = 5

class AccuracyHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.acc = []

    def on_epoch_end(self, batch, logs={}):
        self.acc.append(logs.get('acc'))

history = AccuracyHistory()

```

```python
model.fit(X_train_vals_final, y_train_vals_final,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test_vals_final, y_test_vals_final),
          callbacks=[history])
```

```python
score = model.evaluate(X_test_vals, y_test_vals, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
plt.plot(range(len(history.acc)), history.acc)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
